Q2. An electrical manufacturing company conducts quality checks at specified periods on the products it manufactures. Historically, the failure rate for the manufactured item is 5%. Suppose a random sample of 10 manufactured items is selected. Answer the following questions.

```python
#2.A. Probability that none of the items are defective?

p   =  0.05 # failure rate for the manufactured item
n   =  10 # random sample of 10 manufactured items is selected
k   =  np.arange(0,11) # An array of quantiles

pmf = stats.binom.pmf(k,n,p)
pmf[0]
```

```python
#2.B. Probability that exactly one of the items is defective?

pmf[1]
```

```python
#2.C. Probability that two or fewer of the items are defective?
cdf = stats.binom.cdf(k,n,p)
cdf[2]
```

```python
#2.D. Probability that three or more of the items are defective
P = 1- cdf[2]
P
```

Q3. A car salesman sells on an average 3 cars per week.

```python
rate =  3 # average / mean
n    =  np.arange(0,16)
pcdf = stats.poisson.cdf(n,rate)
pcdf

#3.A. What is Probability that in a given week he will sell some cars? [1 Mark]
P   = 1 - pcdf[0]
P
```

```python
#3.B. What is Probability that in a given week he will sell 2 or more but less than 5 cars? [1 Mark]
# Let P1 be the probability that the salesman sells more than 4 cars
P1 = pcdf[4]
# Let P2 be the probability that the salesman sells more than 1 cars
P2 = pcdf[1]
# Let P be the probability that in a given week he will sell 2 or more but less than 5 cars
P = P1 - P2
P
```

```python
#3.C. Plot the poisson distribution function for cumulative probability of cars sold per-week vs number of cars sold per week. [1 Mark]
poisson = stats.poisson.pmf(n,rate)
plt.plot(n,poisson,)
plt.xlabel('Number of cars sold per week')
plt.ylabel('Cumulative Prob of cars sold per-week')
plt.show()
```

B.  Write the null hypothesis and alternative hypothesis.

Null Hypothesis (H0):
The null hypothesis (H0) is implicitly represented in the code as the statement being tested for significance. It assumes that there is no significant difference in funds raised between the 'Operating' and 'Closed' companies. Specifically, the null hypothesis is:
Alternative Hypothesis (H1):
The alternative hypothesis (H1) is not explicitly written in the code but can be inferred from the context. It suggests that there is a significant difference in funds raised between the 'Operating' and 'Closed' companies. In mathematical notation:

C. Test for significance and conclusion

```python
import pandas as pd
from scipy import stats

# Assuming you have a DataFrame 'company_cleaned' with 'OperatingState' indicating the company status and 'Funds_Million' as the funds raised column
operating_funds = company_cleaned[company_cleaned['OperatingState'] == 'Operating']['Funds_in_million']
closed_funds = company_cleaned[company_cleaned['OperatingState'] == 'Closed']['Funds_in_million']

# Perform the t-test
t_statistic, p_value = stats.ttest_ind(operating_funds, closed_funds, equal_var=False)

# Set the significance level (alpha)
alpha = 0.05

# Interpret the results
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in funds raised between the 'Operating' and 'Closed' companies.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in funds raised between the 'Operating' and 'Closed' companies.")
```
